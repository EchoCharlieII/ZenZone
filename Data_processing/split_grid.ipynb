```python
import geopandas as gpd
import numpy as np
from shapely.geometry import Polygon

# define boundary of Manhattan

manhattan_boundary = Polygon([(-74.0187128181894, 40.70675991106398), (-74.0121973301691, 40.70115009311718), 
                              (-73.97918682110672, 40.710475466843754), (-73.97233649404521, 40.729875043572605),(-73.92991433643208, 40.79508327381935)])

# define column, rows count of grid
num_rows = 20
num_cols = 10

# split grid
x_coords = np.linspace(manhattan_boundary.bounds[0], manhattan_boundary.bounds[2], num_cols + 1)
y_coords = np.linspace(manhattan_boundary.bounds[1], manhattan_boundary.bounds[3], num_rows + 1)

polygons = []
grid_ids = []
for row in range(num_rows):
    for col in range(num_cols):
        xmin, xmax = x_coords[col], x_coords[col + 1]
        ymin, ymax = y_coords[row], y_coords[row + 1]
        polygon = Polygon([(xmin, ymin), (xmax, ymin), (xmax, ymax), (xmin, ymax)])
        polygons.append(polygon)
        grid_ids.append(row * num_cols + col + 1)

# create GeoDataFrame of grid
grid = gpd.GeoDataFrame({'grid_id': grid_ids, 'geometry': polygons})


print(grid['geometry'][1])
```
